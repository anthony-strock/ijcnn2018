```python
scenario = "mult_many_trigger_producthp"
id_res = 0
path_internals = "data/{:s}/0_internals.npy".format(scenario)
internals = np.load(path_internals, mmap_mode='r')[id_res]
path_components = "data/{:s}/0_components.npy".format(scenario)
components = np.load(path_components, mmap_mode='r')[id_res]
path_abs_eigs = "data/{:s}/0_abs_eigs.npy".format(scenario)
abs_eigs = np.load(path_abs_eigs, mmap_mode='r')[id_res]
path_outputs = "data/{:s}/0_outputs.npy".format(scenario)
outputs = np.load(path_outputs,mmap_mode='r')[id_res]
path_inputs = "data/{:s}/0_inputs.npy".format(scenario)
inputs = np.load(path_inputs, mmap_mode='r')
id_seq = 0
```

```python
i = 0
j = 2
f = scatter_2d_all_runs_color(components[:1,:,i], -components[:1,:,j], outputs[:1,:,0], xlabel = "PC{:d}".format(i+1), ylabel = "PC{:d}".format(j+1), figsize = (5,5), s = 5, clabel = "Memory", alpha = 1)
f.set_tight_layout(True)
f.savefig("img/figure6right.pdf",transparent = True)
```

```python
i = 10
j = 32
cimin,cimax = np.min(components[0,:,i]),np.max(components[0,:,i])
cjmin,cjmax = np.min(components[0,:,j]),np.max(components[0,:,j])
omin,omax = np.min(outputs[0,:,0]),np.max(outputs[0,:,0])
idx = np.where(inputs[0,:,1]==1)[0]
ci = [components[0,:idx[0],i]]
cj = [components[0,:idx[0],j]]
o = [outputs[0,:idx[0],0]]

for a in range(1,len(idx)):
    ci.append(components[0,idx[a-1]:idx[a],i])
    cj.append(components[0,idx[a-1]:idx[a],j])
    o.append(outputs[0,idx[a-1]:idx[a],0])

ci.append(components[0,idx[-1]:,i])
cj.append(components[0,idx[-1]:,j])
o.append(outputs[0,idx[-1]:,0])

f = plot_2d_all_runs_color(ci, cj, o, xlabel = "PC{:d}".format(i+1), ylabel = "PC{:d}".format(j+1), figsize = (5,5), clabel = "Memory", clim = [omin,omax], xlim = [cimin,cimax], ylim = [cjmin,cjmax])

f.set_tight_layout(True)
f.savefig("img/figure6left.pdf", transparent = True)
f.show()
```
